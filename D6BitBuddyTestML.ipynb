```python
#Create some placeholders for values that contain average heartbeat at low end and high end depending on Age
low_range = 0
high_range = 0

if (age == '0'):
    low_range = 100
    high_range = 170
if (age == '1'):
    low_range = 95
    high_range =162
if (age == '2'):
    low_range = 93
    high_range =157
if (age == '3'):
    low_range = 90
    high_range = 153
if (age == '4'):
    low_range = 88
    high_range = 149
if (age == '5'):
    low_range = 85
    high_range = 145
if (age == '6'):
    low_range = 83
    high_range = 140
if (age == '7'):
    low_range = 80
    high_range = 136
if (age == '8'):
    low_range = 78
    high_range = 132
if (age == '9'):
    low_range = 75
    high_range = 128
if (age == '10'):
    low_range = 73
    high_range = 125
print(low_range)
print(high_range)
print(high_range-low_range)
```

```python
#Now let's create some data which makes this unrealistic in some way but we are doing a Low-Fidelity Prototype
#First lets create an empty array to hold some values
HBPM = []
#Heart Beat Per Minute or HBPM will hold some Heart Beat monitering 
for x in range (0, 100):
    HBPM.append(random.randint(low_range, low_range+15))

#Let it hold from low with some buffer
for x in range (0, 100):
    HBPM.append(random.randint(low_range, low_range+30))

#Now lets get some random of full range
for x in range (0, 100):
    HBPM.append(random.randint(low_range, high_range))

#Now let us produce some values out of range, near end for debugging purposes
for x in range (0, 100):
    HBPM.append(random.randint(low_range-15, high_range+15))

print(len(HBPM))
```

```python
#Let us simply run through the program and just report range outreach
for i in HBPM:
    if (i < low_range):
        print("Heart Beat getting Low")
    elif (i > high_range):
        print("Heart Beat Getting Higher")
    else:
        continue
    
```

```python
#So we can easily parse and understand if data is out of range but let us include small dependencies
for i in range (0,400):
    if(i == 0):
        print("Test")
    elif (HBPM[i] < HBPM[i-1]):
        print("Heart Beat began to decrease")
    elif (HBPM[i] > HBPM[i-1]):
        print("Heart Beat began to increase")
    else:
        continue
```

```python
#Now to have a certain specific dependency 
Threshold = 3
for i in range (0,400):
    if(i == 0):
        print("Test")
    elif ((HBPM[i-1]-HBPM[i])>Threshold ):
        print(i)
        print("Heart Beat decreased quite quickly")
    elif ((HBPM[i] - HBPM[i-1]) > Threshold):
        print(i)
        print("Heart Beat increased quite quickly")
    else:
        continue
```
